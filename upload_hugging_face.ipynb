### Notebook contain code to make a final JSON to be uploaded to Hugging Face
    
    
    -- Code flow 
        
        -- Read the Meta JSONS made in the first notebook
        
        -- Go to each JSON which contain discourse about individual discourse 
        
        -- Each Row of the final dataset contains
    
            -- Charcater URL under which the discourse falls 
            -- Topic Name of the discourse
            -- Lesson name of the discourse under that Topic
            -- URL of the discourse
            -- Complete text of the URL
    
    -- Total number of rows is 1965

```python
import json
```

```python
from pprint import pprint
```

```python
f = open('index.json')
load_index = json.load(f)
```

```python
pprint(load_index)
```

```python
inv_load_index = {}
for k,v in load_index.items():
    inv_load_index[v]=k
```

```python
pprint(inv_load_index)
```

```python
f = open('data.json')
load_meta = json.load(f)
```

```python
pprint(load_meta)
```

```python
make_full_list = []
for char_url, meta_1 in load_meta.items():
    for each_topic_meta in meta_1:
        for topic_name, topic_meta in each_topic_meta.items():
            for topic_lesson_name, topic_lesson_url in topic_meta.items():
                this_tuple = [char_url, topic_name, topic_lesson_name, topic_lesson_url]
                make_full_list.append(this_tuple)
```

```python
pprint(make_full_list, indent=5)
```

```python
import os
```

```python
cwd = os.getcwd()
cwd
```

```python
base_dir = "files_new_1"
final_dir = os.path.join(cwd, base_dir)
final_dir
```

```python
all_files = os.listdir(base_dir)
pprint(len(all_files))
```

```python
final_list = []
```

```python
for char_url, topic_name, topic_lesson_name, topic_lesson_url in make_full_list:
    file_name = inv_load_index[topic_lesson_url] + '.txt'
    if file_name in all_files:
        with open(os.path.join(final_dir, file_name), 'r', encoding='utf-8') as f:
            all_txt = f.read()
            final_list.append([char_url, topic_name, topic_lesson_name, topic_lesson_url, all_txt])
```

```python
#pprint(final_list)
```

```python
final_dict = {}

char_url_list = []
topic_name_list = []
topic_lesson_name_list = []
topic_lesson_url_list = []
all_txt_list = []
```

```python
for char_url, topic_name, topic_lesson_name, topic_lesson_url, all_txt in final_list:
    char_url_list.append(char_url)
    topic_name_list.append(topic_name)
    topic_lesson_name_list.append(topic_lesson_name)
    topic_lesson_url_list.append(topic_lesson_url)
    all_txt_list.append(all_txt)
```

```python
final_dict['char_url'] = char_url_list
final_dict['topic_name'] = topic_name_list
final_dict['topic_lesson_name'] = topic_lesson_name_list
final_dict['topic_lesson_url'] = topic_lesson_url_list
final_dict['all_txt'] = all_txt_list
```

```python
final_data = {}
final_data['version'] = 2.0
final_data['data'] = final_dict
```

```python
final_json_name = 'ds_hf_2.json' 
```

```python
with open(final_json_name, 'w', encoding='utf-8') as f:
    json.dump(final_data, f, ensure_ascii=False, indent=4)
```
